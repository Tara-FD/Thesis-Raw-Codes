```python
import numpy as np
import pandas as pd
import os, shutil
```

```python
df = pd.read_csv('df_with_all_BN_Schaefer100(functionals).csv')
df.head()
```

```python
src = '/media/tara/HD710PRO/Thesis/FreesurferOutput100'
src
```

```python
s = os.listdir(src)
```

```python
df2 = df.set_index('Subject_ID')
for i,folder1 in enumerate(s):
    path1 = src + '/' + folder1
    ls = os.listdir(path1)
    for folder2 in ls:
        dest = path1 + '/' + folder2
        fso = pd.read_csv(dest + '/parc-stats/' + 'parc_nodes.csv')
        columns = fso.columns[5:7]
        strucIDs = fso['structureID']
        for strucID in strucIDs:
            for col in columns:
                measurement = strucID + '_' + col
                value = fso[fso['structureID'] == strucID][[col]].values[0][0] 
                if measurement not in df2:
                    df2.insert(7, measurement, np.nan, allow_duplicates=False) 
                df2.loc[s[i].split('.')[0],measurement] = value 
```
