```python
import pandas as pd
import os
import shutil
```

```python
fd_cor = pd.read_csv('/media/tara/HD710PRO/Thesis/group_bold_mcrr.csv')
```

```python
#fd_cor = fd_cor.drop('Unnamed: 0', axis=1)
fd_cor.head()
```

### spliting subjects and waves from bids_name

```python
fd_cor["bids_name"] = fd_cor["bids_name"].str.split("_")
```

```python
fd_cor.insert(2, "subject", "0")
fd_cor.insert(3, "wave", "0")
fd_cor = fd_cor.drop('Unnamed: 0', axis=1)
fd_cor.head()
```

```python
for i in range(fd_cor.shape[0]):
    fd_cor["subject"][i] = fd_cor["bids_name"][i][0]
    fd_cor["wave"][i] = fd_cor["bids_name"][i][1]
```

```python
fd_cor.head()
```

### deleting ses-wave1 rows

```python
w2 = fd_cor["wave"] == "ses-wave2"
w3 = fd_cor["wave"] == "ses-wave3"
fd_cor = fd_cor[w2 | w3]
```

```python
#fd_cor = fd_cor.set_index('subject')
fd_cor.shape[0]
```

```python
subs= fd_cor[['subject', 'wave']]
subs.head()
```

### deleting waves of subs with fd<0.5

```python
directory = '/media/tara/HD710PRO/Thesis/NICAP_ADHD_sleep_DATA_FD'

for i, sub in enumerate(subs['subject'].transpose()):
    try:
        wave = subs['wave'][i]
        path = directory + '/' + sub + '/' + wave
        shutil.rmtree(path)
    except:
        print(sub)
        continue
```

### deleting empty folders

```python
c=-1
for file in os.listdir(directory):
    path = directory + '/' + file
    ss=os.listdir(path)
    for i in range(len(ss)):
        if ss[i].find('.')==0:
            continue
        c+=1
    if c==-1:
        shutil.rmtree(path) # Delete..
    c=-1
```
