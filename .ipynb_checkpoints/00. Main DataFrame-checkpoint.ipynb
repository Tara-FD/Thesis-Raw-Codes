### spliting imaging waves + deleting wave1s + deleting SWSless

```python
df = pd.DataFrame(columns=main.columns)
main2 = main
```

```python
main2["imaging waves"] = main2["imaging waves"].str.split(",")
main2["SWS-waves"] = main2["SWS-waves"].str.split(",")
main2.head()
```

```python
main2 = main2.dropna(axis=0, subset=['SWS-waves'])
index = main2['Subject_ID']
main2 = main2.set_index('Subject_ID')
```

```python
for i in index:
    a = main2.loc[[i]]
    s = a['imaging waves']
    sws = a['SWS-waves'][0]
    a = a.reset_index()
    for j in range(len(s[0])):
        if s[0][j] == '1':
            continue
        try:
            if int(sws[0])-2 == int(s[0][j]):
                a['SWS-waves']=1
            elif int(sws[1])-2 == int(s[0][j]):
                a['SWS-waves']=1
            else:
                a['SWS-waves']=0
        except:
            a['SWS-waves']=0
        a['imaging waves'] = 'ses-wave'+str(s[0][j])
        df = pd.concat([df,a])
```

```python
dfsws = df[df['SWS-waves']==1].reset_index(drop=True)
```

```python
inds = dfsws['Subject_ID'].drop_duplicates()
dfsws.shape
```

### merging dfsws with group_bold(fd) = dfgb

```python
dfgb = dfsws.merge(group_bold, on=['Subject_ID' , 'imaging waves'], how='left')
dfgb.shape
```
