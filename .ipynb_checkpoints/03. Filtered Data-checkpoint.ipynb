### spliting subjects and waves from bids_name

```python
mriqc["bids_name"] = mriqc["bids_name"].str.split("_")
```

```python
mriqc.insert(1, "subject", "0")
mriqc.insert(2, "wave", "0")
mriqc.head()
```

```python
for i in range(mriqc.shape[0]):
    mriqc["subject"][i] = mriqc["bids_name"][i][0]
    mriqc["wave"][i] = mriqc["bids_name"][i][1]
```

```python
mriqc.head()
```

### deleting ses-wave1

```python
w2 = mriqc["wave"] == "ses-wave2"
w3 = mriqc["wave"] == "ses-wave3"
mriqc2 = mriqc[w2 | w3]
```

```python
mriqc2.head()
```

```python
mriqc2.drop('Unnamed: 0', axis=1)
```

### keeping min(fd_mean)

```python
index = mriqc2["subject"].drop_duplicates()
```

```python
mriqc2 = mriqc2.set_index("subject")
```

```python
index
```

```python
df=pd.DataFrame(columns=mriqc.columns)
for file in index:
    sub = mriqc2.loc[[file]]
    if sub.shape[0] > 1:
        a=sub['fd_mean'].min()
        s=sub[sub['fd_mean']==a]
    else:
        s=sub
    s=s.reset_index()
    df=pd.concat([df, s])
```

```python
df=df.drop('Unnamed: 0', axis=1)
```

```python
df.head()
```
