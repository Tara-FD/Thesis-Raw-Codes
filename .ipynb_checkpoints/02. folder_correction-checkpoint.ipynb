### Bye Bye (empty folders)

```python
c=-1
for file in os.listdir(directory):
    path = directory + '/' + file
    ss=os.listdir(path)
    for i in range(len(ss)):
        if ss[i].find('.')==0:
            continue
        c+=1
    if c==-1:
        shutil.rmtree(path) # Delete..
    c=-1
```
