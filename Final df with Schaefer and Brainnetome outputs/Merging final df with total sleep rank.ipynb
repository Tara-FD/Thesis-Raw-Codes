### choosing rank of the selected wave

```python
maindf = pd.read_csv('final_df.csv')
maindf = maindf.drop('Unnamed: 0', axis=1)
maindf.head()
```

```python
try:
    maindf.set_index('Subject_ID', inplace=True)
except:
     print('Index have been setted')
maindf.head()
inds = maindf.index
```

```python
df = pd.DataFrame()
if 'sws_total' not in maindf:
    maindf.insert(2, 'sws_total', np.nan)
for ind in inds:
    sub = maindf.loc[[ind]]
    if sub['imaging waves'].str.contains('2').values[0]:
        sub['sws_total'] = sub['c4_sws_total']
    elif sub['imaging waves'].str.contains('3').values[0]:
        sub['sws_total'] = sub['c5_sws_total']
    df = pd.concat([df, sub])
```

```python
df.head()
```

```python
df.to_csv("finaldf_swstotal.csv")
```
