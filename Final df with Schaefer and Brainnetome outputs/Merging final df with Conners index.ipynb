```python
maindf = pd.read_csv('merged_df.csv')
maindf = maindf.drop('Unnamed: 0', axis=1)
maindf.head()
```

```python
try:
    maindf.set_index('Subject_ID', inplace=True)
except:
     print('Index have been setted')
inds = maindf.index
maindf.head()
```

```python
for col in maindf.columns:
    print(col)
```

```python
df1 = pd.DataFrame()
if 'cpadhdi_raw' not in maindf:
    maindf.insert(2, 'cpadhdi_raw', np.nan)
for ind in inds:
    sub = maindf.loc[[ind]]
    if sub['imaging waves'].str.contains('2').values[0]:
        sub['cpadhdi_raw'] = sub['p4_cpadhdi_raw_x']
    elif sub['imaging waves'].str.contains('3').values[0]:
        sub['cpadhdi_raw'] = sub['p5_cpadhdi_raw_x']
    df1 = pd.concat([df1, sub])
```

```python
df1.head()
```

```python
df1.to_csv('df_cpadhd_raw.csv')
```

```python
merged = pd.merge(left=df1, right=dfnicap[['id_number', 'p4_cpadhdi_raw', 'p5_cpadhdi_raw']], how='left', left_on='Subject_ID', right_on='id_number')
merged
```

```python
merged.to_csv('df_conners.csv')
```
