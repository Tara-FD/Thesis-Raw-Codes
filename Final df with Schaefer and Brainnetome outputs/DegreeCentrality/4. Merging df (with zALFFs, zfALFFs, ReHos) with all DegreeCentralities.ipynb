```python
df = pd.read_csv('df_with_schaefer_brainnetome_zReHo_CorticalRemoved.csv')
df.head()
```

```python
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

## adding zDC_bin of BN to dataframe

### list of zDC_bin (BN)

```python
# opening the text file
lis = []
with open('zDC_binarized_BN_meants.txt','r') as ALFFout:
   
    # reading each line    
    for line in ALFFout:
   
        # reading each word        
        for ALFF in line.split():
            lis.append(float(ALFF))
            # displaying the words           
            #print(ALFF) 
```

### list of region names in BN Atlas (246parc)

```python
token = open('BN_Atlas_246_LUT.txt','r')
linestoken=token.readlines()
tokens_column_number = 1
parname=[]
for x in linestoken:
    parname.append(x.split()[tokens_column_number])
token.close()
print(parname)
```

```python
len(parname)
```

### merging regions and their values with dataframe

```python
test = df.set_index('Subject_ID')
inds = test.index
test.head()
```

```python
df = pd.DataFrame()
for j, ind in enumerate(inds):
    sub = test.loc[[ind]]
    for i, pars in enumerate(parname):
        col_name = 'zDCbin_'+str(pars)
        if col_name not in sub:
            sub.insert(2, col_name, np.nan, True)
        sub[col_name] = lis[i+j*246]
    df = pd.concat([df, sub])
```

```python
df.head()
```

```python
df.to_csv("df_with_zDCbin_brainnetome.csv")
```
