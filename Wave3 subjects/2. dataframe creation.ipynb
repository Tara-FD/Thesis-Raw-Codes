```python
df = pd.DataFrame(columns=initial_df.columns)
initialdf = initial_df
```

### spliting imaging waves

```python
initialdf["imaging waves"] = initialdf["imaging waves"].str.split(",")
initialdf["SWS-waves"] = initialdf["SWS-waves"].str.split(",")
initialdf.head()
```

### deleting swsless

```python
initialdf = initialdf.dropna(axis=0, subset=['SWS-waves'])
index = initialdf['Subject_ID']
initialdf = initialdf.set_index('Subject_ID')
```

```python
for i in index:
    a = initialdf.loc[[i]]
    s = a['imaging waves']
    sws = a['SWS-waves'][0]
    a = a.reset_index()
    for j in range(len(s[0])):
        if s[0][j] == '1':
            continue
        if s[0][j] == '2':
            continue
        
        try:
            if int(sws[0])-2 == int(s[0][j]):
                a['SWS-waves']=1
            elif int(sws[1])-2 == int(s[0][j]):
                a['SWS-waves']=1
            else:
                a['SWS-waves']=0
        except:
            a['SWS-waves']=0
        a['imaging waves'] = 'ses-wave'+str(s[0][j])
        df = pd.concat([df,a])
```
