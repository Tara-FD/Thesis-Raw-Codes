```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os 
import shutil
```

```python
initial_df = pd.read_csv('dataframe_wave3_subjects.csv')
initial_df.head()
```

```python
df = pd.DataFrame(columns=initial_df.columns)
initialdf = initial_df
```

### spliting imaging waves

```python
initialdf["imaging waves"] = initialdf["imaging waves"].str.split(",")
initialdf["SWS-waves"] = initialdf["SWS-waves"].str.split(",")
initialdf.head()
```

### deleting swsless

```python
initialdf = initialdf.dropna(axis=0, subset=['SWS-waves'])
initialdf.head()
```

```python
index = initialdf['Subject_ID']
main2 = initialdf.set_index('Subject_ID')
index
```

```python
main2.to_csv(r'/media/tara/HD710PRO/main2.tsv', index = False)
```

```python
main2 = pd.read_csv('main2.tsv')
#main2 = 
```

```python
for i in index:
    a = main2.loc[[i]]
    s = a['imaging waves']
    sws = a['SWS-waves'][0]
    a = a.reset_index()
    for j in range(len(s[0])):
        if s[0][j] == '1':
            continue
        if s[0][j] == '2':
            continue
        try:
            if int(sws[0])-2 == int(s[0][j]):
                a['SWS-waves']=1
            elif int(sws[1])-2 == int(s[0][j]):
                a['SWS-waves']=1
            else:
                a['SWS-waves']=0
        except:
            a['SWS-waves']=0
        a['imaging waves'] = 'ses-wave'+str(s[0][j])
        df = pd.concat([df,a])
```
