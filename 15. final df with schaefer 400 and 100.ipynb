## adding schaefer100 to dataframe

### list of ALFFs

```python
# opening the text file
lis = []
with open('allzALFF_meants.txt','r') as ALFFout:
   
    # reading each line    
    for line in ALFFout:
   
        # reading each word        
        for ALFF in line.split():
            lis.append(float(ALFF))
            # displaying the words           
            #print(ALFF) 
```

```python
len(lis)
```

### list of region names in Schaefer Atlas (100 parc-17 net)

```python
token = open('Schaefer2018_100Parcels_17Networks_order.txt','r')
linestoken=token.readlines()
tokens_column_number = 1
parname=[]
for x in linestoken:
    parname.append(x.split()[tokens_column_number])
token.close()
print(parname)
```

```python
newdf = df.copy(deep = True)
```

```python
test = newdf.set_index('Subject_ID')
inds = test.index
# test to df2
```

```python
df = pd.DataFrame()
for j, ind in enumerate(inds):
    sub = test.loc[[ind]]
    for i, pars in enumerate(parname):
        if pars not in sub:
            sub.insert(2, pars, np.nan, True)
        sub[pars] = lis[i+j*100]
    df = pd.concat([df, sub])
```

```python
df = df.reset_index()
```
