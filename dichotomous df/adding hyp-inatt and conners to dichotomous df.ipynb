```python
import pandas as pd
```

```python
dichdf = pd.read_csv('dichotomus_df.csv')
dichdf = dichdf.drop('Unnamed: 0', axis=1)
dichdf.head()
```

```python
nicapdf = pd.read_csv('NICAP_Data_info.csv')
nicapdf.head()
```

```python
finaldich = pd.merge(left=dichdf, right=nicapdf[['d3_ad_inattcri_y', 'd3_ad_hypcri_y', 'id_number']], how='left', left_on='Subject_ID', right_on='id_number')
finaldich
```

```python
finaldich.to_csv('dichotomous.csv')
```
