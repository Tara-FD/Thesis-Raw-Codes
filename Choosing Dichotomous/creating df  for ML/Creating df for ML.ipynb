```python
df = pd.read_csv('final(All100).csv')
df.head()
```

```python
feat = pd.read_csv('bonferroni_reformed.csv')
feat = feat[['Unnamed: 0', 'zfALFF_sws_total_bonferroni']]
feat = feat.dropna()
feat = feat.reset_index()
feat = feat.drop('index', axis = 1)
feat
```

```python
df2 = pd.DataFrame()
df2 = pd.concat([df2, df['Subject_ID']])
for f in feat["Unnamed: 0"]:
    f = 'zfALFF_' + f
    for i,column in enumerate(df.columns):
        if f == column:        
            #print(f)
            df2 = pd.concat([df2, df[[column]]], axis=1)
            break
```
