```python
import pandas as pd
import os
import shutil
import numpy as np
```

### uploading and cleaning dataframe

```python
df = pd.read_csv('ALL_DF_400.csv')
df = df.drop('Unnamed: 0', axis = True)
df.head()
```

```python
for col in df.columns:
    print(col)
```

### removing ALFF and DCbin

```python
df_sub1 = df[df.columns.drop(list(df.filter(regex='zDCbin')))]
df_sub1.shape
```

```python
df_sub2 = df_sub1[df_sub1.columns.drop(list(df_sub1.filter(regex='zALFF')))]
df_sub2.shape
```

### adding age gender and adhd status

```python
info = pd.read_csv('info.csv')
info = info.drop('Unnamed: 0', axis=1)

info.head()
```

```python
merged = pd.merge(left=df_sub2,right=info, how='left', left_on='Subject_ID', right_on='Subject_ID')  
merged.shape
```

```python
merged.drop('CHILDgender', axis = 1, inplace=True)
merged.insert(1, 'CHILDgender', info['CHILDgender'])
```

```python
merged.drop('diagadhd', axis = 1, inplace=True)
merged.insert(1, 'diagadhd', info['diagadhd'])
```

```python
merged.head()
```

```python
age = pd.read_csv('age.csv')
age.head()
```

```python
merged.insert(2, 'age', age['age'])
merged.head()
```

```python
merged.drop(['sws_gb', 'sws_ms', 'sws_fa', 'sws_rw', 'sws_rw'], axis = 1, inplace=True)
```

```python
merged.drop('sws_rs', axis = 1, inplace=True)
merged.head()
```

```python
df = merged
```

### reordering columns based on parcels and indexes

```python
df.head()
```

```python
#if len(df.columns) == 
# df.columns = df.columns.str.replace("17Networks_1", "17Networks_01")
```

```python
columns = df.columns[5:]
```

```python
#lst = ["_".join(c.split("_")[1:]) for c in columns.values]
```

```python
ls = []
for cols in columns:
    #cols.split('_')
    mod = cols.split("_")[1:]
    par = ["_".join(cols.split("_")[1:])]
    #par = par.sort_values()
    #print(par)
    ls.append(par)
```

```python
pars = np.sort(ls, axis = 0)
```

```python
for par in pars:
    print(par)
```

```python
ordered_pars = list()
for item in pars:
    if item not in ordered_pars:
        ordered_pars.append(item[0])
ordered_pars
```

```python
len(ordered_pars)
```

```python
order = []
for l in ordered_pars:
    i1 = 'gmd_' + l
    i2 = 'zfALFF_' + l
    i3 = 'zDCwei_' + l
    i4 = 'zReHo_' + l
    order.append(i1)
    order.append(i2)
    order.append(i3)
    order.append(i4)
    
```

```python
order
```

```python
df_ordered = df[order]
```

```python
df_ordered.head()
```

```python
df_ordered['Subject_ID'] = df['Subject_ID']
```

```python
df_ordered.drop('Subject_ID', axis = 1, inplace=True)
df_ordered.insert(0, 'Subject_ID', df['Subject_ID'])
```

```python
df_ordered.insert(1, 'diagadhd', df['diagadhd'])
```

```python
df_ordered.insert(2, 'CHILDgender', df['CHILDgender'])
```

```python
df_ordered.insert(3, 'age', df['age'])
```

```python
df_ordered.insert(4, 'sws_total', df['sws_total'])
```

```python
df_ordered
```

```python
df_ordered.to_csv('finaldf_orderd.csv')
```
