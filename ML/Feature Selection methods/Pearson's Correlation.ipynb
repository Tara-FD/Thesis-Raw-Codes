```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('df_184feat.csv')
df.drop('Unnamed: 0', axis=1, inplace=True)
df.drop('Unnamed: 0.1', axis=1, inplace=True)
df.head()
```

```python
df = df.drop(columns = ['Subject_ID', 'diagadhd', 'CHILDgender'])
df.head()
```

```python
vals = df.drop('sws_total', axis=1)
```

```python
from scipy.stats import pearsonr

# Convert dataframe into series
list1 = df['sws_total']
list2 = vals.values
 
# Apply the pearsonr()
corr, _ = pearsonr(list1, list2)
print('Pearsons correlation: %.3f' % corr)
```
