```python
import pandas as pd
import numpy as np
import shutil,os
```

```python
r = pd.read_csv('SWSms_r_reformed.csv')
r.head()
#r.shape
```

```python
p = pd.read_csv('SWSms_p_reformed.csv')
p = p[["parc_name", "zALFF_sws_ms_p", "zDCbin_sws_ms_p", "zDCwei_sws_ms_p", "zfALFF_sws_ms_p", "zReHo_sws_ms_p", "gmd_sws_ms_p"]]
p.head()
```

```python
p.shape
```

```python
r.shape
```

```python
r_p = r.merge(p, how = 'left', left_on = 'parc_name', right_on = 'parc_name')
```

```python
r_p.shape
```

```python
r_p.head()
```

```python
fdr = pd.read_csv('SWSms_fdr_reformed.csv')
fdr = fdr[["parc_name", "zALFF_sws_ms_fdr", "zDCbin_sws_ms_fdr", "zDCwei_sws_ms_fdr", "zfALFF_sws_ms_fdr", "zReHo_sws_ms_fdr", "gmd_sws_ms_fdr"]]

fdr.head()
```

```python
fdr.shape
```

```python
r_p_fdr = r_p.merge(fdr, how = 'left', left_on = 'parc_name', right_on = 'parc_name')
```

```python
r_p_fdr.shape
```

```python
bonferroni = pd.read_csv('SWSms_bonferroni_reformed.csv')
bonferroni = bonferroni[["parc_name","zReHo_sws_ms_bonferroni"]]
bonferroni.head()
```

```python
bonferroni.shape
```

```python
r_p_fdr_bonf = r_p_fdr.merge(bonferroni, how = 'left', left_on = 'parc_name', right_on = 'parc_name')
```

```python
df = r_p_fdr_bonf
```

```python
df.head()
```

```python
df2 = df.reindex(sorted(df.columns), axis=1)
```

```python
df2 = df2.replace({'0':np.nan, 0:np.nan})
```
