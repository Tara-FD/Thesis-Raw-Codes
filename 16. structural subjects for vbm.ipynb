```python
for file in os.listdir(dir):
    path = dir + '/' + file
    s1 = os.listdir(path)
    finalpath = os.path.join(path, s1[-1]) + '/anat'
    s2 = os.listdir(finalpath) 
    for s2i in s2:
        if s2i.find('.')==0:
            continue
        if s2i.find('T1w.') != -1:
            original = finalpath + '/' + s2i
            target = '/media/tara/HD710PRO/Thesis/NICAP_ADHD_sleep_DATA(freesurfer_design)/' + file + '/anat' 
            os.makedirs(target, exist_ok = True)
            target = target + '/' + s2i
            shutil.copyfile(original, target)
```
