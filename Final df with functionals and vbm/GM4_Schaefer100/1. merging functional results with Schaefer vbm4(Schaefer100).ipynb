```python
import numpy as np
import pandas as pd
import os, shutil
```

```python
df = pd.read_csv('df_with_all_BN_Schaefer100(functionals).csv')
df.head()
```

### list of vbm(Schaefer100)

```python
# opening the text file
lis = []
with open('GM4_Schaefer100_meants.txt','r') as ALFFout:
   
    # reading each line    
    for line in ALFFout:
   
        # reading each word        
        for ALFF in line.split():
            lis.append(float(ALFF))
            # displaying the words           
            #print(ALFF)
```

```python
len(lis)
```

### list of region names in Schaefer Atlas(100 parc)

```python
token = open('Schaefer2018_100Parcels_17Networks_order.txt','r')
linestoken=token.readlines()
tokens_column_number = 1
parname=[]
for x in linestoken:
    parname.append(x.split()[tokens_column_number])
token.close()
print(parname)
```

### merging regions and their values with dataframe

```python
test = df.set_index('Subject_ID')
inds = test.index
```

```python
df = pd.DataFrame()
for j, ind in enumerate(inds):
    sub = test.loc[[ind]]
    for i, pars in enumerate(parname):
        col_name = 'gmd_'+str(pars)
        if col_name not in sub:
            sub.insert(2, col_name, np.nan, True)
        sub[col_name] = lis[i+j*100]
    df = pd.concat([df, sub])
```
