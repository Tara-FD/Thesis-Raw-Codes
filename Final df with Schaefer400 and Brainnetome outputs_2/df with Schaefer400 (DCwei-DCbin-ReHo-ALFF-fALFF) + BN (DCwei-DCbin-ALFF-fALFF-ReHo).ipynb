```python
import numpy as np
import pandas as pd
import os, shutil
```

```python
df = pd.read_csv('df with Schaefer400_DCwei-DCbin-ReHo-ALFF-fALFF_and_BN DCwei-DCbin-ALFF-fALFF.csv')
df.head()
```

```python
# opening the text file
lis = []
with open('allzReHo_meants_BN.txt','r') as ALFFout:
   
    # reading each line    
    for line in ALFFout:
   
        # reading each word        
        for ALFF in line.split():
            lis.append(float(ALFF))
            # displaying the words           
            #print(ALFF) 
```

```python
token = open('BN_Atlas_246_LUT.txt','r')
linestoken=token.readlines()
tokens_column_number = 1
parname=[]
for x in linestoken:
    parname.append(x.split()[tokens_column_number])
token.close()
print(parname)
```

```python
test = df.set_index('Subject_ID')
inds = test.index
```

```python
df = pd.DataFrame()
for j, ind in enumerate(inds):
    sub = test.loc[[ind]]
    for i, pars in enumerate(parname):
        col_name = 'zReHo_'+str(pars)
        if col_name not in sub:
            sub.insert(2, col_name, np.nan, True)
        sub[col_name] = lis[i+j*246]
    df = pd.concat([df, sub])
```
